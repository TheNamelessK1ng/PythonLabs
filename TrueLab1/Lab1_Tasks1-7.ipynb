## Завдання 1
### Створіть функцію, яка приймає назву файлу та назву стовпця як параметр і повертає масив numpy зі значеннями стовпця.

```python
import numpy as np
import csv

def get_equipment(file_name, col_title):
    list = []
    with open(file_name, mode='r') as file:
        csvFile = csv.reader(file)
        header_row = next(csvFile)
        col_index = header_row.index(col_title)
        for row in csvFile:
            if col_title != 'date':
                list.append(int(row[col_index]))
            else:
                list.append(str(row[col_index]))
    return np.array(list)
```

```python
get_equipment('russia_losses_equipment.csv', 'tank')
```

## Завдання 3
### Знайдіть добові втрати танків.

```python
import numpy as np
import csv

def find_num_of_tank_losses_per_day(tanks):
    days_list = []
    with open('russia_losses_equipment.csv') as file:
        csvFile = csv.reader(file)
        header_row = next(csvFile)
        col_index = header_row.index('day')
        for row in csvFile:
            days_list.append(int(row[col_index]))
        days = np.array(days_list)
        daily_losses = np.abs(np.diff(tanks))
        for day, losses in zip(days[1:], daily_losses):
            print(f"Day {day}, Tanks lost: {losses}")
```

```python
find_num_of_tank_losses_per_day(tanks)
```

## Завдання 4
### Знайдіть 4 найбільші добові втрати танків.

```python
import numpy as np

def find_biggest_daily_losses(num, tanks):
    daily_losses = np.abs(np.diff(tanks))
    dates = np.argsort(daily_losses)
    max_dates = dates[-num:]
    return daily_losses[max_dates], max_dates + 1
```

```python
losses, days = find_biggest_daily_losses(4, tanks)
print("Найбільші втрати:", losses)
print("Дні, коли вони сталися:", days)
```

## Завдання 5
### Визначте, скільки танків було знищено навесні 2023 року.

```python
import numpy as np
from datetime import datetime
import csv

def find_spring_losses(start, end):
    start_date = datetime.strptime(start, '%Y-%m-%d')
    end_date = datetime.strptime(end, '%Y-%m-%d')
    total_tanks = []
    with open('russia_losses_equipment.csv') as file:
        csvFile = csv.reader(file)
        header_row = next(csvFile)
        date_col_index = header_row.index('date')
        tank_col_index = header_row.index('tank')
        for row in csvFile:
            row_date = datetime.strptime(row[date_col_index], '%Y-%m-%d')
            if (start_date <= row_date <= end_date):
                total_tanks.append(int(row[tank_col_index]))
        total_losses = np.diff(total_tanks)
        return np.abs(np.sum(total_losses))
```

```python
print(f"Losses on spring 2023:", find_spring_losses('2023-03-01', '2023-05-31'))
```

## Завдання 6
### Знайдіть середнє значення знищених танків за останні 100 днів війни.

```python
import numpy as np
import csv

def find_average_value_tanks(tanks_array):
    count = 0
    result = 0
    daily_losses = np.diff(tanks_array)
    for element in daily_losses:
        if count == 100:
            break
        result += element
        count += 1
    if count > 0:
        return np.abs(result / count)
    else:
        return 0
```

```python
tanks = get_equipment('russia_losses_equipment.csv', 'tank')
find_average_value_tanks(tanks)
```

## Завдання 7
### Створіть графік втрат танків за минулий рік (починаючи з останньої дати в наборі даних)

```python
import numpy as np
import matplotlib.pyplot as plt

def get_dates_and_tanks(file_name, dates_title, tanks_title, start_date, end_date):
    start = datetime.strptime(start_date, '%Y-%m-%d')
    end = datetime.strptime(end_date, '%Y-%m-%d')
    total_tanks = []
    total_dates = []
    with open(file_name) as file:
        csvFile = csv.reader(file)
        header_row = next(csvFile)
        date_col_index = header_row.index(dates_title)
        tank_col_index = header_row.index(tanks_title)
        for row in csvFile:
            row_date = datetime.strptime(row[date_col_index], '%Y-%m-%d')
            if (start <= row_date <= end):
                total_tanks.append(int(row[tank_col_index]))
                total_dates.append(str(row[date_col_index]))
        total_tanks = total_tanks[::-1]
        total_dates = total_dates[::-1]
        tank_losses = np.abs(np.diff(total_tanks))
        dates = np.array(total_dates[:-1])
    return tank_losses, dates

def build_graph():
    losses, dates = get_dates_and_tanks('russia_losses_equipment.csv', 'date', 'tank', '2023-01-01', '2023-12-31')
    xAxis = dates
    yAxis = losses
    plt.figure(figsize=(20, 6), dpi=100)
    plt.title("Tank Losses Over The Last Year")
    plt.xlabel("Date")
    plt.ylabel("Number of Tanks Lost")
    ticks_position = np.arange(0, len(xAxis), 14)
    plt.xticks(ticks_position, labels=xAxis[ticks_position], rotation=45)
    plt.plot(xAxis, yAxis, marker='o', linestyle='--', color="green")
    plt.grid(True, color="orange")
    plt.savefig("tank_losses_over_the_last_year.png", format='png', dpi=75)
    plt.show()
```

```python
build_graph()
```
